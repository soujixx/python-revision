```python
name -> name of the file
mode-> which mode file is
closed-> return boolean value
readable -> return boolean if the file is readable
writable -> return boolean value if the file is writable or in write mode
```

```python
f = open("test.txt" ,"w")
```

```python
print(f.name)
print(f.mode)
print(f.closed) # here the file is not closed
print(f.readable())
print(f.writable())
f.close() # here the file is closed coz we're closing the file 
print(f.closed) #  so we get True cox actually the file is closed here 
```
